### 3.4.1 線形予測子と対数リンク関数

ポアソン回帰によるサイズに応じた種子数の統計モデルを探す（肥料効果は一旦無視）

### 3.4.1　線形予測子と対数リンク関数

個体$i$の種子数$y_i$がポアソン分布に従うとして、
\begin{eqnarray}
p(y_i|\lambda_i)=\frac{\lambda_i^{y_i}\exp(-\lambda_i)}{y_i}
\end{eqnarray}
この$\lambda_i$をサイズ$x_i$の関数として定義する。
\begin{eqnarray}
\lambda_i=\exp(\beta_1+\beta_2x_i)
\end{eqnarray}
ただし、ここで$\beta_1$: 切片、$\beta_2$: 傾きである。


```python
def f_lam(x,b1,b2):
    return np.exp(b1+b2*x)

xi=np.arange(-5,5,0.1)
plt.plot(xi,f_lam(xi,-2,-0.8), 'r--', label=r'$\beta_1=-2$, $\beta_2=-0.8$')
plt.plot(xi,f_lam(xi,-1,0.4), 'r-', label=r'$\beta_1=-1$, $\beta_2=0.4$')
plt.xlim([-4.2,4.5])
plt.axvline(x=0, color='k',ls='--',lw=0.2)
plt.xlabel(r'Size: $x_i$')
plt.ylabel(r'Mean: $\lambda_i$')
plt.legend()
```

この時、
\begin{eqnarray}
\log\lambda_i=\beta_1+\beta_2x_i
\end{eqnarray}
となり、右辺は線形予測子と呼ばれる。（{$\beta_1,\beta_2,\beta_3, \cdots$}の線形結合)   
また、左辺の関数をリンク関数と呼び、この場合($\log$)対数リンク関数と呼ぶ。

正準関数(GLMのモデルでよく使われる関数)
* ポアソン回帰->対数リンク関数
* ロジスティック関数->ロジットリンク関数


#### ポアソン回帰

観測データに対するポアソン分布を使った統計モデルの当てはめ  
本例題の場合、$\beta_{1,2}$の推定値を対数尤度の最大値から求めること
\begin{eqnarray}
\log{L(\beta_1,\beta_2)}=\Sigma_i\log\frac{\lambda_i^{y_i}\exp(-\lambda_i)}{y_i!}
\end{eqnarray}

  
たいていの場合解析的には解けないので、数値的に求める

### Ex.1) statsmodels

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
#model = smf.glm('y ~ x', data=data3a_pd, family=sm.families.Poisson())
model = smf.glm('y ~ x', data=data3a_pd, family=sm.families.Poisson()) #family: 分布関数の指定
#model2 = smf.glm('y ~ x + I(x**2)', data=data3a_pd, family=sm.families.Poisson()) #family: 分布関数の指定
results = model.fit()
#results2=model2.fit()
print(results.summary())

#print("\n        \n")

#print(results2.summary())
```

P>|z|は大きいほどz値が小さい、つまり推定値がzeroであることを表現する  
この場合、$\beta_1$については意味がなさそう

### Ex.2) scipy.stats.lineregress (線形フィット)

```python
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)

print 'slope', slope
print 'intercept', intercept
print 'r value', r_value
print  'p_value', p_value
print 'standard deviation', std_err


line = slope*x+intercept
plt.plot(x,y,'o')
plt.plot(x,line,'r--', label='Linear')
plt.plot(x,np.exp(1.2917+0.0757*x),'g--', label='Poission') #Poisson回帰
plt.legend(loc='best')
```
