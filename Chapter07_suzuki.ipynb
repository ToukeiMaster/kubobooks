## Chapter7 「一般化線形混合モデル(GLMM)」

単純なGLMでは実際のデータ解析には応用できない
：測定できない or 測定しなかった個体差によって、現実のデータにみられるばらつきはポアソン分布や二項分布だけでは上手く説明できない。

その個体差を組み込んだのが一般化線形混合モデル
＝複数の確率分布を用いる

　データのばらつき：ポアソン or 二項　分布で、個体のばらつき：正規分布でとかいうようにする。


◎ 過分散と個体差

実際のデータは、例えば二項分布で期待されるよりも大きなばらつきを持つ ⇒ 過分散

この過分散は、観測されていない個体差がひとつの要因であると考えられている。

個体差と言うのは、具体的には

・遺伝子や年齢、過去に経験した履歴の差といった生物的なもの

・生育環境の差といった非生物的なもの　（場所差）

が考えられる。

これらの個体差・場所差は全て定量化することはできないので、これらの影響をうまくとりこんだ統計モデルをつくるということが重要になる。



### 7.3 一般化線形混合model

個体差の効果を考慮したロジスティック回帰のGLMM

個体$i$の個体差を表すパラメータ$r_i$を追加する。

logit($q_i$) = $\beta _1$ + $\beta _2$ $x_i$ + $r_i$

$r_i$ : -$\infty$ - $\infty$ の範囲をとる連続値


個体差のばらつきをどう表すか??

100個の個体の個体差{r1, r2, ..., r100} が何かの確率分布に従っていると仮定する。
(実際に$r$がどのような確率分布を持つのかは分からない。)

⇒ ex )  平均ゼロ、標準偏差s の正規分布
          $p(r_i|s) = \frac{1}{\sqrt(2\pi s^2)} {\rm exp}(-r^2_i/2s^2)$
 
 

 この確率密度関数は、$r_i$の出現のしやすさを表していると解釈できる。$r_i$=0付近の個体は「ありがち」。
 標準偏差sは、集団内の$r_i$のばらつきを表していて、sが大きいほどその集団の個体差が大きく、過分散も大きくなりそう。
 
 
GLMMの線形予測子は、
logit($q_i$) = $\beta _1$ + $\beta _2$ + $r_i$
と表したが、

$\beta_1$, $\beta_2 x_i$ を固定効果、

個体差$r_i$をランダム効果

と呼ぶ。
固定効果は全体の平均を変えて、ランダム効果は平均は変化させないけど全体のばらつきを変える、という感じ。


このふたつの要素で線形予測子が表されているので、「混合」モデルと呼ばれる。



### 7.4 一般化線形混合モデルの最尤推定

各個体の個体差$r_i$を含めて最尤推定することはできない(パラメータが多すぎる)。

⇒　個体ごとの尤度$L_i$の式の中で$r_i$を積分する。

$L_i = \int^{\infty}_{-\infty} p(y_i|\beta_1, \beta_2, r_i)p(r_i|s)dr_i$

⇒ いろいろな$r_i$をとる時の尤度を評価し、$p(r_i|s)$で重み付けした上で、その期待値を求めていることになる。

：2種類の確率分布を混ぜている。

そうすると全データの尤度は

$L(\beta_1, \beta_2, s) = \Pi Li$


となって、この対数尤度が最大となる様なパラメータ$\beta_1, \beta_2, s$を探す。

GLMMによってパラメータ推定を行った結果、

logit($q_i$) = $\beta _1$ + $\beta _2$$x_i$ + $r_i$

としたとき、パラメータの最尤推定値は、
$\hat\beta _1 = -4.190$, $\hat \beta _2 = 1.005$ となり、これらは真の値と近い値になっている。

また、$r_i$に関しては、Scale parameter で標準偏差$s$の最尤推定値が与えられている。
Error は$s$の推定値の誤差。

$s=2.408$ (真の値は3らしく、過小評価気味？)

AICは、275.4と得られた。


```python
# GLMMから得られた葉数(x)と生存種子数(y)の関係
aa = np.array([0.5, 1., 1.5, 2., 2.5, 3., 3.5, 4., 4.5, 5., 5.5, 6., 6.5, 7.], dtype=float)
bb = np.empty(14)
dd = np.empty(14)
dd2 = np.empty(14)

for ii in range(0,14):
 bb[ii] = -4.190 + 1.005 * aa[ii]  #logit(qi) with r=0
    
 dd[ii] = 1.0/(1.0+math.exp(-bb[ii]))
    
 dd2[ii] = 1.0/(1.0+math.exp(-(-2.1487+0.5104*aa[ii]))) 
    
 #print aa[ii], bb[ii], 8.0*dd[ii]

#plot
plt.scatter(data_pd["x"], data_pd["y"], color="red")
plt.plot(aa, 8.0*dd2, color="green", linestyle="--")
plt.plot(aa, 8.0*dd, color="blue")
# 後者のほうが上手く観測データを再現できている。。

xlim(1.9,6.1)

plt.show()
```

### 7.5 現実のデータ解析にはGLMMが必要

どういった場合にGLMMを用いる必要があるのか？

判断のポイント?

- 同じ個体・場所などから何度もサンプリングしているか？
- 個体差や場所差を識別できてしまうようなデータのとりかたをしているか？


◎ 反復・疑似反復と統計モデルの関係

・反復

１個体からひとつの種子をとってきてその生死を調べる。とかひとつの植木鉢には１個体しかない。というように

データの取り方が「反復」である場合には、GLMで推定をしても問題ない。

⇒　個体差・場所差を識別できない。

・疑似反復

個体から複数のデータをとっているような場合は疑似反復となって、パラメータの推定の際にはGLMMを用いて個体差を組み込む必要がある。


統計モデリングを行う時には、データがどのようにしてとられたのかということが重要となる。
