#### 9.2 GLMのベイズモデル化



これまで通りのPoisson回帰により、
個体$i$の種子数$y_i$は平均$\lambda_i$のPoisson分布で表現できるとすると、
\begin{equation}
p(y_i)=p_P(y_i|\lambda_i)
\end{equation}
Poission分布の正準リンク関数である対数リンク関数を用いて
\begin{equation}
\log{\lambda_i}=\beta_1+\beta_2x_i 
\end{equation}
と表現する。(ランダム効果無し)  
このとき、ある $\beta_1, \beta_2$ が与えられた時の尤度関数は

\begin{equation}
L(\beta_1,\beta_2)=\prod_ip_P(y_i|\lambda_i)=\prod_ip_P(y_i|\beta_i,\beta_2,x_i)
\end{equation}

また、ここで8章のBayes統計モデルの考え方を用いると、あるデータ*$Y$*が与えられた時の推定パラメータの事後分布は
\begin{equation}
p(\beta_1,\beta_2|Y) \propto p(Y | \beta_1,\beta_2) p(\beta_1) p(\beta_2)
\end{equation}
となる。
- $p(\beta_1,\beta_2|Y)$: 事後分布
- $p(Y | \beta_1,\beta_2) = L(\beta_1,\beta_2)$
- $p(\beta_{1,2})$: prior

-> **このpriorを適切に設定する事でBayesモデル化したGLMが可能**  
(ただし、priorの設定は非常に気を使わなければいけない)


####9.3 無情報事前分布
本来は、何かしら意味のある事前情報を入れて適切なパラメータ推定を行うのが  
Bayes統計モデルの長所だが、ここでは**推定パラメータについて何の事前情報も得られていない**状況を考える。  
(今考えている架空植物の種子数を表現するパラメータについて何も！知らない！！)  
-> **無情報事前分布(non-informative prior)**

このような場合、priorの入れ方は二種類ある

- 完全な一様分布($1e-9<\beta_*<1e+9、$のように有限の範囲で)
- 平べったい正規分布(ある程度$\beta_*$の範囲がわかっている必要が有るが)

以後は、二番目の平べったい一様分布を用いる


####9.4 Bayes統計モデルの事後分布の推定

- このsectionはツールの使い方ばかりで、さらにPython環境で対応できなかったので中身だけざっくり紹介します。m(\_ \_)m

#### 8章と同様のmetlopolis法で9章の例題のMCMC推定をやってみる

```python
from scipy.stats import poisson, norm

#Flat prior
def logp1(b1_proposed):
    mu=0.0
    sigma=100.0
    return np.log(norm.pdf(b1_proposed, mu, sigma))

def logp2(b2_proposed):
    mu=0.0
    sigma=100.0
    return np.log(norm.pdf(b2_proposed, mu, sigma))

def metropolis_bayes(ns,b10,db1,b20,db2,size,Nseed): #sample size, b1_0, db1, b2_0, db2, size, Nseed

    nstep = np.arange(ns)
    b1, b2 = np.zeros(ns), np.zeros(ns)
    b1_old, b2_old=b10, b20 #初期値    
    
    mu_old = np.exp(b1_old + b2_old * size)
    logl_old = np.sum(poisson.logpmf(Nseed, mu_old))
    logl_p_old = logl_old + logp1(b1_old) + logp2(b2_old) # const.はムシ

    for i in nstep:
        ####b1 sampling###
        b1_new = b1_old + np.random.choice([-db1,db1]) 
        #b2_new = b2_old + np.random.choice([-db2,db2])

        mu_new = np.exp(b1_new + b2_old * size)
        #mu_new = np.exp(b1_new + b2_new * size)
        logl_new = np.sum(poisson.logpmf(Nseed, mu_new))
        logl_p_new = logl_new + logp1(b1_new) + logp2(b2_old) # const.はムシ
        #logl_p_new = logl_new + logp1(b1_new) + logp2(b2_new) # const.はムシ

    
        cr = np.random.rand()
        p_trans = np.exp(logl_p_new - logl_p_old)
        if(cr < p_trans): 
            b1[i] = b1_new
            #b2[i] = b2_new
            logl_p_old = logl_p_new
            b1_old = b1_new
            #b2_old = b2_new
        else:
            b1[i] = b1_old
            #b2[i] = b2_old
            
        ####b2 sampling###    
        b2_new = b2_old + np.random.choice([-db2,db2])
        mu_new = np.exp(b1[i] + b2_new * size)
        logl_new = np.sum(poisson.logpmf(Nseed, mu_new))
        logl_p_new = logl_new + logp1(b1[i]) + logp2(b2_new) # const.はムシ
    

        cr2 = np.random.rand()
        p_trans2 = np.exp(logl_p_new - logl_p_old)
        if(cr2 < p_trans2): 
            b2[i] = b2_new
            logl_p_old = logl_p_new
            b2_old = b2_new
        else:
            b2[i] = b2_old
       
            
    sns.jointplot(x=b1, y=b2)
    sns.jointplot(x=b1, y=b2, kind='kde', color='orangered')
    sns.jointplot(nstep,b1,xlim=(0,ns), color='g')
    sns.jointplot(nstep,b2,xlim=(0,ns), color='y')
    
    return b1, b2
    
```

```python
size, Nseed = np.array(data9.x), np.array(data9.y)
#Flat prior (test, prior off)
metropolis_bayes(3000,0.0,0.02,0.0,0.02,size,Nseed) #b10=0.0, b20=0.0
```

```python
#Flat prior (prior off)
metropolis_bayes(50000,1.5,0.02,0.1,0.02,size,Nseed) #b10=1.5, b20=0.1
```

$\beta_1\sim1.9$, $\beta_2\sim0.05$あたりに収束しているように見える

初期値を変えても、同じところに収束している($N_{step}\gtrsim 5000$)

```python
#Prior on (実際のbeta1, beta2の*正しい*分布を知っている場合)
def logp1(b1_proposed):
    mu=1.5
    sigma=0.1
    return np.log(norm.pdf(b1_proposed, mu, sigma))

def logp2(b2_proposed):
    mu=0.1
    sigma=0.1
    return np.log(norm.pdf(b2_proposed, mu, sigma))
metropolis_bayes(10000,2.0,0.02,0.5,0.02,size,Nseed) #b10=2.0, b20=0.5
```

この場合、事後分布はflat priorの場合と近く**収束が早い**   
-> priorを正しく設定していれば、ベイズ統計モデルは有効

```python
#Prior on (beta1, beta2の*誤った*事前分布を入れてしまった場合)
def logp1(b1_proposed):
    mu=3.0
    sigma=0.1
    return np.log(norm.pdf(b1_proposed, mu, sigma))

def logp2(b2_proposed):
    mu=1.0
    sigma=0.1
    return np.log(norm.pdf(b2_proposed, mu, sigma))
metropolis_bayes(10000,2.0,0.02,0.5,0.02,size,Nseed)
```

この場合、おかしなところで収束してしまう  
->やはりpriorは魔法の手段だけど、リスクもある！！
