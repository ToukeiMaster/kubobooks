### statsmodels のパラメータの渡し方
上記のdocumentによると、  

    The statistical model is assumed to be
\begin{equation}
   Y = X\beta + \mu, \mathrm{where}\ \mu\sim N\left(0,\sigma^{2}\Sigma\right)
\end{equation}

    depending on the assumption on $\Sigma$, we have currently four classes available

    GLS : generalized least squares for arbitrary covariance $\Sigma$  
    OLS : ordinary least squares for i.i.d. errors $\Sigma=\textbf{I}$  
    WLS : weighted least squares for heteroskedastic errors $\text{diag}\left  (\Sigma\right)  $  
    GLSAR : feasible generalized least squares with autocorrelated AR(p) errors $\Sigma=\Sigma\left(\rho\right)$
   

とあるので、*線形予測子は一次式が前提*っぽいです

###4.1 

前章で一般線形化モデル(GLM)を導入して架空植物の種子数$y_i\ (i=1,2,...,100)$を説明する統計モデルを作って（Poisson回帰）R or Pythonで最尤推定をした。  
図4.2は各ポアソン回帰モデルを示すが、果たしてどのモデルが*「良い」*モデルだろうか。  
ここで、もし「**最大対数尤度:$Log(L^*)$** が最も大きいモデルこそが最も優れたモデルである」と言う考え方はよくよく考えれば正しくないと分かる。  
つまり、フィットする統計モデルを複雑にすれば「当てはまりの良いモデル」はいくらでも作れてしまう。  
極端なことを言えばデータ数の分だけ変数を準備してデータ点を必ず通過するような**フルモデル**を作れば、これが最も当てはまりの良いモデルとなる。

###4.2: 逸脱度(deviance)

\begin{equation}
D=-2\log{L^*} ~ (\log{L^*}:\mathrm{maximum\ log\ likelihood})
\end{equation}

で定義される「当てはまりの悪さ」を表す量  
"-1"でなく"-2"をかけるのは5章の $\chi^2$ 分布との対応が良くなるため、らしい

```python
results.deviance #statsmodelsの出力のdeviance
```

ここで出力されているdevianceは、上で定義した逸脱度を使って  
*D - min. deviance*  (最小逸脱度)  
で定義された量**（残差逸脱度）**である。  
最小逸脱度はフルモデルにおける逸脱度のこと

フルモデルの最大対数尤度:
\begin{equation}
 \log{L_{max}}=\log{L(\lambda_i=y_i)}=\Sigma_i\log\frac{y_i^{y_i}\exp{(-y_i)}}{y_i!}
\end{equation}

残差逸脱度はフルモデルでの逸脱度に対する「当てはまりの悪さ」の相対値、と言える  
下に示すように、モデルのパラメータ数を増やせば残差逸脱度は小さくなる

##### フルモデルの（最小）逸脱度を求める

```python
def maxdev(x):
    return sct.poisson.logpmf(x,x) 

logpmf_max=np.sum(maxdev(y)) # lambda_i=y_iの時(フルモデル)の最大対数尤度
deviance_min=-2*logpmf_max # D_min=logL_max * (-2)
print deviance_min # 最小逸脱度
```

このときのx+fモデルの残差逸脱度は上で求めた最小逸脱度を用いて

```python
print results.llf *(-2) - deviance_min #定義通りに計算
print results.deviance #statsmodels
```

->statsmodelsの出力の(residual) devianceと一致した！

```python
print "llf:%f, residual deviance: %f" % (results2.llf, results2.deviance)
```

##### 6次式モデルの残差逸脱度

### 4.3 AIC

何らかの基準の基に良いモデルを選択する事: **モデル選択**  
代表的なモデル選択規準の一つである**AIC (Akaike's information criterion)**を導入する   
これは、「予測の良さ(*goodness of prediction*)」を重視するモデル選択基準である

(参考)https://ja.wikipedia.org/wiki/赤池情報量規準


#### AICの定義

\begin{eqnarray}
AIC&=&-2(\log{L^*}-k)\\
&=&D+2k\\
\end{eqnarray}
ここで、$k$は最尤推定したパラメータの数  

この時、**AICが最も小さくなるようなモデルが良いモデル**となる  
表4.3は種子数を表現する各統計モデルに対してAICを載せている

- null: k=1, logL*=-237.6, residual deviance=89.5, AIC=477.3
- f   : k=2, logL*=-237.6, residual deviance=89.5, AIC=479.3
- x   : k=2, logL*=-235.4, residual deviance=85.0, AIC=474.8
- x+f : k=3, logL*=-235.3, residual deviance=84.8, AIC=476.6
- full: k=100, logL*=-192.9, residual deviance=0.0, AIC=585.8

->AICを基にすると**"xモデル"**がAIC最小で「最も予測が良いモデル」  
なぜAIC最小が良いのかは、この後の章で。


### 4.4以降

以下の問題を検証する
- AICで選ばれたモデルは何が「良い」のか
- AICの定義はどこから来ているのか

######この後の内容をまとめると...  
結論としては、ある観測データから得られた統計モデルの予測の良さを表す**「平均対数尤度 (mean log likelihood)」**の推定量をAICは表している  
解析的、かつ一般的に平均対数尤度の推定量が$\log{L^*}-k$であると知られていることから上記のようにAICが定義されている（詳細は文献参照）

###4.5.1　平均対数尤度
- **最大対数尤度$\log{L^*}$はある（たまたま得られた）観測データに対する当てはまりの良さを表している**にすぎない(図4.6)    
    - なので通常の真の統計モデルが不明である場合に、この値に固執して何とかより当てはまりの良い統計モデルを探すのはナンセンス  
    - 実際のデータ解析では、こういった発想をしがち
- 正しいアプローチは**「推定した統計モデル」が「真の統計モデル」にどれだけ近いかを評価すること**
    - ここで、試しに真の統計モデルがわかっている($\beta_1=2.08$)として、そこから得られる観測データを生成
    - そこから推定される統計モデルの「予測の良さ」を評価してみる
    - 「予測の良さ」...**次に**同じデータ取得方法で別のデータを得た時に、**始め**のデータから推定した統計モデルがどれくらい正確に当てはまるか
- そこで登場するのが**平均対数尤度 E(log*L*)**
    - (例)200個の観測データセット(それぞれ50個の種子数を測定)から、各データセットに対して一度一つのデータセットから推定したnull model($\hat{\beta}_1=2.04$)の当てはまりの良さを対数尤度で評価し、その平均をとる（図4.8）

###4.5.2　最大対数尤度のバイアス補正

- 最大対数尤度と平均対数尤度の関係（図4.9(A)）
    - $\log{L^*}=-120.6$
    - $E(\log{L})=-122.9$  
    
つまり、ある一つの観測データから推定した統計モデルは当てはまりの良さを**過剰評価**している  
ただし、200個全て過剰評価している、ということではなくあくまで**平均として**過剰評価（図4.9(A)）

この操作を12回繰り返したのが図4.9(B)

最大対数尤度と平均対数尤度の差を
\begin{equation}
b=\log{L^*}-E(\log{L})
\end{equation}
と定義する。これを**バイアス**とよぶ。  
図4.9(C)は12回のバイアスの分布  
図4.10は同じ操作を200回繰り返した時のbの分布であり、標本平均1.01  
つまり、
\begin{equation}
E(\log{L}) \sim \log{L^*} - 1.01
\end{equation}

平均対数尤度は統計モデルの良さの指標として有効だが、
この量は真の統計モデルがわかっていないと知る事ができない。

そこで、再び上の式を見てみると、
**平均対数尤度が最大対数尤度からバイアスを引いた値として推定できる**、とわかる**（バイアス補正）**  
なので、最大対数尤度$\log{L^*}$とバイアス$b$の値がわかれば良い。

####で、どうやって$b$を求めるかと言うと...

- 「さて、数理統計学の教科書を参照してみると、最尤推定するパラメータをk個もつモデルの平均対数尤度の推定量は$\log{L^*}-k$である--といったことが解析的かつ一般的に導出されています。」\\(^o^)/


ということで、結局
\begin{equation}
AIC=-2\times(\log{L^*}-k)\sim -2E(\log{L})
\end{equation}

**-> AICは平均対数尤度$E\log{L}$の推定値$\log{L^*}-k$に-2をかけたもの**であり、   
AICで統計モデルを選択することは複数の統計モデルのAICを比較して**「予測の悪さ」がより小さいモデルを選ぶこと**に相当する
